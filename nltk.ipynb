```python
import nltk
import spacy
import pandas as pd 
import numpy as np
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv('NER_Dataset.csv')
data = df
```

```python
sentences =  data['Word'].apply(eval).apply(' '.join)
```

```python
nltkpos_listoflists = []
for sentence in sentences:
    text = nltk.pos_tag(word_tokenize(sentence))
    #only extracting the tag from the tuple
    nltkpos_list = [x[1] for x in text]
    nltkpos_listoflists.append(nltkpos_list)
```

```python
tp2 = data.loc[:,'POS']
tp2_listoflists = tp2.apply(eval).tolist()
tp2_listoflists
```

```python
import itertools
pred_nltkpos = list(itertools.chain(*nltkpos_listoflists))
true_nltkpos = list(itertools.chain(*tp2_listoflists))
```

```python
u3 = set(pred_nltkpos)
u4 = set(true_nltkpos)
```

```python
for i in u3:
    if i not in u4:
        print(i)
print('-------------------')
for i in u4:
    if i not in u3:
        print(i)
```

```python
pred_nltkposctr = pd.Series(pred_nltkpos).value_counts()
true_nltkposctr = pd.Series(true_nltkpos).value_counts()
```

```python
diff = {}
total = 0
for i in u4:
    if i in u3:
        diff[i] = pred_nltkposctr[i] - true_nltkposctr[i]
        total += abs(diff[i])
    if i not in u3:
        total += true_nltkposctr[i]

accuracy = 1 - total/len(pred_nltkpos)
accuracy
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
pos2 = []
xyz = []
# Step Three: Tokenise, find parts of speech and chunk words 
for sentence in sentences:
    for sent in nltk.sent_tokenize(sentence):
        temp = nltk.pos_tag(nltk.word_tokenize(sent))
        pos2.append(temp)
        for chunk in nltk.ne_chunk(temp):
            
            if hasattr(chunk, 'label'):
                xyz.append(chunk.label())
```

```python
nt = data.loc[:,'Tag']
nt_listoflists = nt.apply(eval).tolist()
ner_true = nt_listoflists
ner_truemerged2 = list(itertools.chain(*ner_true))
ner_truemerged = []
for i in ner_truemerged2:
    if i != 'O':
        ner_truemerged.append(i)
```

```python
len(xyz)
```

```python
ner_truemergedctr = pd.Series(ner_truemerged).value_counts().to_dict()
ner_predmergedctr = pd.Series(xyz).value_counts().to_dict()

ner_truemergedctr['B-geo'] = ner_truemergedctr['B-gpe'] + ner_truemergedctr['B-geo']
del ner_truemergedctr['B-gpe']
print(ner_truemergedctr)
print(ner_predmergedctr)
```

```python
# Second dictionary provided
second_dict = ner_predmergedctr

#accoording to https://www.ldc.upenn.edu/sites/www.ldc.upenn.edu/files/edt-guidelines-v2-5.pdf

# Mapping from second to original keys with only B- labels
reverse_mapping = {
    #'DATE': 'B-tim',
    #'TIME': 'B-tim',
    'PERSON': 'B-per',
    'GPE': 'B-geo',
    'ORGANIZATION': 'B-org',
    'LOCATION': 'B-eve',
    'GSP': 'B-nat',
    'FACILITY': 'B-art',
    #'WORK_OF_ART': 'B-art',
    #'EVENT': 'B-eve'
}

# New dictionary to hold converted values
converted_dict = {}

# Convert without summing as we're using only B- labels
for key, value in second_dict.items():
    new_key = reverse_mapping.get(key)  # Get the new key from the mapping
    if new_key:  # If the key exists in the mapping
        converted_dict[new_key] = value

print(converted_dict)
```

```python
diff = {}
total = 0
for i in ner_truemergedctr:
    if i in converted_dict:
        diff[i] = converted_dict[i] - ner_truemergedctr[i]
        total += abs(diff[i])
    # if i not in converted_dict:
    #     total += ner_truemergedctr[i]

accuracy = 1 - total/len(ner_truemerged)
accuracy
```

```python
print("confusion matrix \n matches: ",len(ner_truemerged) - total, "\n total: ", len(ner_truemerged), "\n accuracy: ", accuracy)
```
